```python
import pandas as pd
import folium
```

```python
location = "JC-202106-citibike-tripdata.csv"
```

```python
bike_station_locations = pd.read_csv(location)
```

```python
bike_station_locations
```

```python
bike_station_locations.head(1000)
```

```python
bike_station_locations1 = bike_station_locations[["start_lat", "start_lng", "start_station_name", "memb_casu"]]
```

```python
bike_station_locations1
```

```python
bike_sample = bike_station_locations1.head(2000)
```

```python
# Creating the map and adding points to it
map = folium.Map(location=[bike_sample.start_lat.mean(), bike_sample.start_lng.mean()], zoom_start=14, control_scale=True)

for index, location_info in bike_sample.iterrows():
    folium.Marker([location_info["start_lat"], location_info["start_lng"]], popup=location_info["start_station_name"]).add_to(map)
```

```python
map
```

```python
from folium.plugins import HeatMap

map1 = folium.Map(location=[bike_sample.start_lat.mean(), bike_sample.start_lng.mean()], zoom_start=14, control_scale=True)

HeatMap(data=list(zip(bike_sample['start_lat'], bike_sample['start_lng'], bike_sample['memb_casu'])),radius = 35).add_to(map1)

map1
```
